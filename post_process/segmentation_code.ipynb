```python
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d.axes3d as p3 
import numpy as np
from matplotlib import animation
from mpl_toolkits.mplot3d import Axes3D 
from scipy.integrate import solve_ivp 
import cv2
%matplotlib notebook
```

```python
class DroneMask():
    def __init__(self):
        self.kernel = np.ones((3,3),np.uint8)

    def getmask(self, img):
        img = cv2.cvtColor(cv2.imread(img, cv2.IMREAD_COLOR),cv2.COLOR_BGR2RGB)
        hsvImg = cv2.cvtColor(img,cv2.COLOR_RGB2HSV)

        #multiple by a factor to change the saturation
        hsvImg[...,1] = hsvImg[...,1]

        #multiple by a factor of less than 1 to reduce the brightness 
        hsvImg[...,2] = hsvImg[...,2]
        
        
        lower_gray = np.array([0,0,90])
        upper_gray = np.array([360,30,250])
        gray = cv2.inRange(hsvImg, lower_gray,upper_gray)
        

        mask = gray
        mask[mask==0]=1
        mask[mask==255]=0
        dilated_drone = cv2.dilate(mask,self.kernel, iterations=2)

        return img, dilated_drone
        
```
